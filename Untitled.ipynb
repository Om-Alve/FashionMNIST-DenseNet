```python
import torch 
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST,FashionMNIST
from utils import train_transforms,calculate_accuracy_and_loss,test_transforms
from config import *
from model import Classifier
from tqdm import tqdm
import torch.nn.functional as F
from torch.utils.tensorboard import SummaryWriter
import numpy as np
```

```python
train_dataset = FashionMNIST(root='~/datasets',train=True,transform=train_transforms,download=True)
```

```python
BATCH_SIZE = 512
train_dataset = FashionMNIST(root='~/datasets',train=True,transform=train_transforms,download=True)
test_dataset = FashionMNIST(root='~/datasets',train=False,transform=test_transforms,download=True)
train_dataloader = DataLoader(train_dataset,batch_size=256,shuffle=True,num_workers=4)
test_dataloader = DataLoader(test_dataset,batch_size=BATCH_SIZE,shuffle=False,num_workers=2)
```

```python
num_epochs = 2
```

```python
model = Classifier().to(DEVICE)
opt = torch.optim.AdamW(lr=3e-4,params=model.parameters())
totalsteps = num_epochs * len(train_dataloader)
r = np.linspace(start=-4,stop=-1,num=(totalsteps))
lrs = 10 ** r
```

```python
losses = []
acc = []
```

```python
torch.set_float32_matmul_precision('high')
step = 0
for i in range(num_epochs):
    for x,y in tqdm(train_dataloader):
        x = x.to(DEVICE)
        y = y.to(DEVICE)
        logits = model(x)
        opt.param_groups[0]['lr'] = lrs[step]
        loss = F.cross_entropy(input=logits,target=y)
        losses.append(loss.item())
        opt.zero_grad()
        loss.backward()
        opt.step()
        step+=1
        preds = logits.argmax(dim=-1).type(torch.long)
        num_correct = (preds == y).sum()
        acc.append(calculate_accuracy_and_loss(test_dataloader,model,DEVICE)[0])
    print(loss.item(),acc[-1])
```

```python
# Plotting the loss vs. learning rate
plt.plot(lrs, acc)
plt.xscale('log')
plt.xlabel('Learning Rate')
plt.ylabel('Loss')
plt.title('Learning Rate Range Test')
plt.show()
```
